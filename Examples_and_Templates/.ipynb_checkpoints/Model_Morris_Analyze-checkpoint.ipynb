### Generate Results for Method of Morris within a Model

This script utilizes simulation results, stored in a pickle file, generated previously.

Utilizing the entire model for the Method of Morris is ideal when considering bulk and wall species.

#### Import Packages

```python
#Import packages:

import os 
import sys

main_folder=r'C:\Users\frank\Documents\Box Sync\working files\Leap-hi\Epanet-MSX\MSXpy\MSXPY'
sys.path.insert(0,main_folder+'\Function_Libraries')
#Set working directory
os.chdir(main_folder)


import pickle
import numpy as np
import matplotlib.pyplot as plt
from SALib.sample import morris as morris_s
from SALib.analyze import morris as morris_a
import pandas as pd
import wntr
import mf_msx_toolkit as mf
```

##### Define Initial Variables

Define the name of the pickle file to be used, as well as the species of interest.

The species of interest is the species concentration for which the values of mu_star and sigma should be calculated. It can be thought of as: "How sensitive is the concentration of species X based on a change in parameter Y"


```python
pickle_file_model='example_morris.pkl'

species='Xb'
```

#### Extract elements from pickle file

```python
#import model results from pickle file
with open(pickle_file_model,'rb') as pfile:
    model_pickle=pickle.load(pfile)

results_model=model_pickle['results']
nodes=model_pickle['nodes']
links=model_pickle['links']
problem=model_pickle['problem']
param_values_model=model_pickle['param_values']
inp_file=model_pickle['inp_file']
```

#### Calculate values of mu_star and sigma for each timestep of the model evaluation

The function MorrisWallEvaluate calculates the values of mu_star and sigma for each timestep of the model for the species of intrest (defined above). The result is a dictionary which contains the results of the method of morris analysis as well as a dictionary containing a timeseries of the water age at each model element, to be used for further analysis.

Note: the water age calculation expects a bulk species called 'AGE' which tracks water age. Examples of implementing this species are in the two included .msx files.

```python
morris_results, age= mf.MorrisWallEvaluate(model_pickle,species)
```

#### Analyze results

Now that the morris_results dictionary has been created, the results can be analyzed in any way suitable for a particular application. Below are a few examples

Plot mu_star versus water age for a particular node in the network

```python
plt.figure()
node='32'
parameter='Kd'
plt.plot(age['node'][node],morris_results['mu_star']['node'][node][parameter],ls='',marker='o')
plt.xlabel('Water Age (hours)')
plt.ylabel('Mu_Star')
plt.title('Mu_star vs water age | Parameter: '+ parameter+ ' | Node: '+node+' | Species: ' + species)
```

#### Aggregate results of mu_star for each parameter

For each model link, the mean value of mu_star for each parameter is calculated. A boxplot is then shown showing the distribution in mean mu_star value across all links for each element.

In this particular implementation, links 10 and 9 are excluded from this analysis.

```python

links.remove('10')
links.remove('9')
parameters=problem['names']
mu_star_link_df=pd.DataFrame(columns=parameters,index=links)
sigma_link_df=pd.DataFrame(columns=parameters,index=links)


for i in range(len(links)):
    mu_star_df=morris_results['mu_star']['link'][links[i]]
    sigma_df=morris_results['sigma']['link'][links[i]]
    
    link_mu_star_row=[]
    link_sigma_row=[]
    
    for k in range(len(parameters)):
        link_mu_star_row.append(mu_star_df[parameters[k]].mean())
        link_sigma_row.append(sigma_df[parameters[k]].mean())
    
    mu_star_link_df.loc[links[i],:]=link_mu_star_row
    sigma_link_df.loc[links[i],:]=link_sigma_row
    
    
fig,ax= plt.subplots()
ax.boxplot(mu_star_link_df)
ax.set_xticklabels(parameters,rotation=45)
ax.set_ylabel('mu_star')
plt.title('Distribution of mean mu_star values for each link | Species: '+species)
```

#### Create a map showing the mean value of mu_star for each link

```python
#Import network to wntr
wn = wntr.network.WaterNetworkModel(inp_file)

#Choose a parameter
parameter='Kb'
plot_var_link=mu_star_link_df[parameter]
wntr.graphics.plot_network(wn,link_attribute=plot_var_link)
plt.title('Mu_star for parameter: '+parameter +' based on species: '+species )
```
